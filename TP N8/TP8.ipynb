&emsp;&emsp;1.4 Para cada hipótesis, ¿cuál es la probabilidad de que el cuarto tornillo extraído sea de cobre?

### 1.4 Probabilidad de que el cuarto tornillo extraído sea de cobre para cada hipótesis

Sea $H_i$ cada hipótesis sobre la caja de tornillos.  
Supongamos que la probabilidad de extraer un tornillo de cobre bajo cada hipótesis $H_i$ es $P(Cu \mid H_i)$.

La probabilidad de que el **cuarto tornillo extraído sea de cobre** bajo cada hipótesis se calcula como:

$
P(\text{4º tornillo = Cu} \mid H_i) = P(Cu \mid H_i)
$

Es decir, dado que los tornillos se extraen de manera independiente, la probabilidad de que el cuarto tornillo sea de cobre es **igual a la probabilidad de extraer un tornillo de cobre bajo esa hipótesis**, independientemente de los tres tornillos anteriores.

Si tenemos, por ejemplo, las siguientes hipótesis y probabilidades:

| Hipótesis $H_i$ | $P(Cu \mid H_i)$ |
|------------------|------------------|
| \(H_1\)          | 0.1              |
| \(H_2\)          | 0.3              |
| \(H_3\)          | 0.5              |
| \(H_4\)          | 0.7              |
| \(H_5\)          | 0.9              |

Entonces:


\begin{align*}
P(\text{4º tornillo = Cu} \mid H_1) &= 0.1 \\
P(\text{4º tornillo = Cu} \mid H_2) &= 0.3 \\
P(\text{4º tornillo = Cu} \mid H_3) &= 0.5 \\
P(\text{4º tornillo = Cu} \mid H_4) &= 0.7 \\
P(\text{4º tornillo = Cu} \mid H_5) &= 0.9
\end{align*}
 
La probabilidad de que el cuarto tornillo sea de cobre depende únicamente de la hipótesis de la caja de tornillos y no de los tornillos ya extraídos, como se logra ver en la gráfica de verosimilitud en funcion de la cantidad de extracciones con reemplazo.
